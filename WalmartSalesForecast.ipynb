```python
# Creating Data audit Report
# Use a general function that returns multiple values
def var_summary(x):
    return pd.Series([x.count(), x.isnull().sum(), x.sum(), x.mean(), x.median(),  x.std(), x.var(), x.std()/x.mean(), x.min(), x.dropna().quantile(0.01), x.dropna().quantile(0.05),x.dropna().quantile(0.10),x.dropna().quantile(0.25),x.dropna().quantile(0.50),x.dropna().quantile(0.75), x.dropna().quantile(0.90),x.dropna().quantile(0.95), x.dropna().quantile(0.99),x.max()], 
                  index=['N', 'NMISS', 'SUM', 'MEAN','MEDIAN', 'STD', 'VAR', 'CV','MIN', 'P1' , 'P5' ,'P10' ,'P25' ,'P50' ,'P75' ,'P90' ,'P95' ,'P99' ,'MAX'])
```

```python
n=num_data.apply(var_summary)

n.to_csv('num.csv')


```

```python

def outliertreatment(x):
    x=x.clip(upper=x.quantile(0.99))
    x=x.clip(lower=x.quantile(0.01))
    return x


def fillmissing(x):
    x=x.fillna(x.median())
    return x




def fillmissingcat(x):
    x=x.fillna(x.mode())
    return x
```

```python

num_data['MarkDown1'] = np.where(num_data.MarkDown1 < 0, 0,num_data.MarkDown1)
num_data['MarkDown2'] = np.where(num_data.MarkDown2 < 0, 0,num_data.MarkDown2)
num_data['MarkDown3'] = np.where(num_data.MarkDown3 < 0, 0,num_data.MarkDown3)
num_data['MarkDown4'] = np.where(num_data.MarkDown4 < 0, 0,num_data.MarkDown4)
num_data['MarkDown5'] = np.where(num_data.MarkDown5 < 0, 0,num_data.MarkDown5)
num_data['Weekly_Sales'] = np.where(num_data.Weekly_Sales < 0, 0,num_data.Weekly_Sales)

num_data=num_data.apply(outliertreatment)

num_data=num_data.apply(fillmissing)


d=num_data.apply(var_summary)

d.to_csv('after.csv')



```

```python
import statsmodels.formula.api as smf

fval=[]
pval=[]
cols=list(final_data.columns.difference(['ln_Weekly_Sales']))
for col in cols:
    f= 'ln_Weekly_Sales ~ '+col
    model=smf.ols(formula=f,data=final_data).fit()
    fval.append(model.fvalue)
    pval.append(model.f_pvalue)
    

fval=pd.Series(fval,name="FScore")
pval=pd.Series(pval,name="PValue")
cols=pd.Series(cols,name="cols")

d=pd.concat([fval,pval,cols],axis=1)


    
```
